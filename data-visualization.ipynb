```python
import imageio
from matplotlib import pyplot as plt
%matplotlib inline
```

### Extracting our Image

```python
testimg = imageio.imread("datasets/blue_jay.jpg")
plt.imshow(testimg)
```

### Check its size

```python
testimg.shape
```

### Creating Color Filters

```python
plt.subplot(131)
plt.imshow( testimg[:,:,0], cmap="Reds")
plt.title("Red channel")
plt.subplot(132)
plt.imshow( testimg[:,:,1], cmap="Greens")
plt.title("Green channel")
plt.subplot(133)
plt.imshow( testimg[:,:,2], cmap="Blues")
plt.title("Blue channel")
```
